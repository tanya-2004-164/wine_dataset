```python
from datasets import load_dataset
```

```python
# Load the WineQT dataset from Hugging Face
dataset = load_dataset("lvwerra/red-wine")
```

```python
# Convert the dataset to a Pandas dataframe
df = pd.DataFrame(dataset["train"])
```

```python
#Handling Missing Values
print(df.isnull().values.any())
print(df.isnull())
```

```python
df = df.dropna()
```

```python
#plotting box plot for each feature
print(df.plot.box(column = 'fixed acidity'))
print(df.plot.box(column = 'volatile acidity'))
print(df.plot.box(column = 'citric acid'))
print(df.plot.box(column = 'residual sugar'))
print(df.plot.box(column = 'chlorides'))
print(df.plot.box(column = 'free sulfur dioxide'))
print(df.plot.box(column = 'total sulfur dioxide'))
print(df.plot.box(column = 'density'))
print(df.plot.box(column = 'pH'))
print(df.plot.box(column = 'sulphates'))
print(df.plot.box(column = 'alcohol'))
print(df.plot.box(column = 'quality'))
```

```python
#using IQR to handle outliers
for item in df.columns:
  Q1 = df[item].quantile(0.25)
  Q3 = df[item].quantile(0.75)
  IQR = Q3 - Q1

  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR

  df = df[(df[item] >= lower_bound) & (df[item] <= upper_bound)]
```

```python
#Change in outliers can now be seen via the boxplots
print(df.plot.box(column = 'fixed acidity'))
print(df.plot.box(column = 'volatile acidity'))
print(df.plot.box(column = 'citric acid'))
print(df.plot.box(column = 'residual sugar'))
print(df.plot.box(column = 'chlorides'))
print(df.plot.box(column = 'free sulfur dioxide'))
print(df.plot.box(column = 'total sulfur dioxide'))
print(df.plot.box(column = 'density'))
print(df.plot.box(column = 'pH'))
print(df.plot.box(column = 'sulphates'))
print(df.plot.box(column = 'alcohol'))
print(df.plot.box(column = 'quality'))
```
